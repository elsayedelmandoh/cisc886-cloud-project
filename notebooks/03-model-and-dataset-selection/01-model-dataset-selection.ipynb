---

> Why TinyLlama is Appropriate

1. **Size**: 1.1B parameters fits within 16GB VRAM (Colab T4) with 4-bit quantization
2. **Hardware**: Feasible for fine-tuning on free Colab + deployment on g4dn.xlarge
3. **Domain Fit**: Pre-trained on 30% code data (Starcoderdata), suitable for coding assistant
4. **Training Data**: Uses Slimpajama (70% natural language) + Starcoderdata (30% code)
5. **Efficiency**: Unsloth provides 2-4x faster training, 70%+ memory reduction
6. **Chat Capability**: Has instruction tuning for conversational use
