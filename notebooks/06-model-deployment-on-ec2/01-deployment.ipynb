## Commands for EC2 Setup

Run these commands on the EC2 instance after SSH connection.

### SSH into EC2

```bash
ssh -i "25xrvl-key.pem" ubuntu@ec2-*.us-east-1.compute.amazonaws.com
```

### Install Ollama

```bash
curl -fsSL https://ollama.com/install.sh | sh
```

### Download Model from S3

```bash
# Create model directory
mkdir -p ~/models
cd ~/models

# Download model from S3
aws s3 cp s3://25xrvl-s3-project/models/25xrvl-tinyllama-codegpt-q4_k_m.gguf .
```

### Create Modelfile

```bash
cat > Modelfile << EOF 
FROM ./my_model/model_final.gguf 
PARAMETER temperature 0.7 
PARAMETER top_p 0.9
SYSTEM You are a coding assistant specialized in Python, JavaScript and Go.
EOF
```

### Clone conversion tool

```bash
git clone https://github.com/ggerganov/llama.cpp
```

### Install dependencies

```bash
pip install gguf sentencepiece numpy transformers torch --break-system-packages
```

### Run conversion

```bash
python3 llama.cpp/convert_hf_to_gguf.py ~/my_model --outfile ~/my_model/model_final.gguf --outtype q8_0
```

### Import Model to Ollama

```bash
ollama create 25xrvl-codegpt -f Modelfile
```

### Test Model

```bash
ollama run codegpt-25xrvl "Write a hello world in Python"
```

### API Curl Test

```bash
curl http://localhost:11434/api/tags
```
