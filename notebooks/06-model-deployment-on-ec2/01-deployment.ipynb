## Commands for EC2 Setup

Run these commands on the EC2 instance after SSH connection.

### SSH into EC2

```bash
ssh -i "25xrvl-key.pem" ubuntu@ec2-*.us-east-1.compute.amazonaws.com
```

### Install Ollama

```bash
curl -fsSL https://ollama.com/install.sh | sh
```

### Download Model from S3

```bash
# Create model directory
mkdir -p ~/models
cd ~/models

# Download model from S3
aws s3 cp s3://25xrvl-s3-project/models/25xrvl-tinyllama-codegpt-q4_k_m.gguf .
```

### Create Modelfile

```bash
cat > Modelfile << EOF
FROM ./25xrvl-tinyllama-codegpt-q4_k_m.gguf
PARAMETER temperature 0.7
PARAMETER top_p 0.9
SYSTEM You are a coding assistant specialized in Python, JavaScript, Java, and Go.
EOF
```

### Import Model to Ollama

```bash
ollama create 25xrvl-codegpt -f Modelfile
```

### Test Model

```bash
ollama run 25xrvl-codegpt "Hello, how are you?"
```

### API Curl Test

```bash
curl http://localhost:11434/api/generate -d '{
  "model": "25xrvl-codegpt",
  "prompt": "Write a hello world in Python",
  "stream": false
}'
```
