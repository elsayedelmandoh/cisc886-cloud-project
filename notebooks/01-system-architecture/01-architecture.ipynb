# Section 1: System Architecture

## Overview

This section documents the system architecture for the cloud-based conversational chatbot deployed on AWS.

---

## System Components

The system consists of:

| Component | Description |
|-----------|-------------|
| VPC | 10.0.0.0/16 - Virtual Private Cloud |
| Public Subnet | 10.0.1.0/24 - EC2 with Ollama + OpenWebUI |
| Private Subnet | 10.0.2.0/24 - EMR cluster |
| Internet Gateway | Connects VPC to internet |
| Security Group | Firewall rules for EC2 |
| S3 Bucket | 25xrvl-s3-project - Data and model storage |
| EMR Cluster | PySpark for data preprocessing |
| EC2 g4dn.xlarge | GPU instance for model serving |
| Ollama | LLM runner |
| OpenWebUI | Web chat interface |

---


## Data Flow

1. **Raw Dataset**: Downloaded from Hugging Face
2. **S3 Storage**: Uploaded to S3 bucket (25xrvl-s3-project/raw/)
3. **EMR Processing**: PySpark filters and cleans data
4. **Colab Fine-tuning**: QLoRA fine-tuning on processed data
5. **Model Export**: GGUF format uploaded to S3 (25xrvl-s3-project/models/)
6. **EC2 Deployment**: Ollama serves the model
7. **User Interface**: OpenWebUI connects to Ollama for chat
